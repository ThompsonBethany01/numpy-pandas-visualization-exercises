```python
import matplotlib.pyplot as plt
# y = 0x + 1
x = range(0, 10)
y = [(n * 0) + 1 for n in x]

plt.plot(x,y)
plt.show()
```

```python
import math
# y = log(x)
x = range(1,10)
y = [math.log(n) for n in x]

plt.plot(x,y)
plt.show()
```

```python
# y = x
x = range(1,10)
y = [n for n in x]

plt.plot(x,y)
plt.show()
```

```python
# y = x * log(x)
x = range(1,10)
y = [(n * (math.log(n))) for n in x]

plt.plot(x,y)
plt.show()
```

```python
# y = x ^ 2
x = range(1,10)
y = [n ** 2 for n in x]

plt.plot(x,y)
plt.show()
```

```python
# y = 2 ^ x
x = range(1,10)
y = [2 ** n for n in x]

plt.plot(x,y)
plt.show()
```

```python
# y = x!
x = range(1,10)
y = [math.factorial(n) for n in x]

plt.plot(x,y)
plt.show()
```

```python
# y = n ^ n
x = range(1,5)
y = [n ** n for n in x]

plt.plot(x,y)
plt.show()
```

# Now that all plots are working correctly, we will combine into one plot...

```python
#"O(1)"
x1 = range(0, 10)
y1 = [(n * 0) + 1 for n in x1]

#"O(log n)"
x2 = range(1,10)
y2 = [math.log(n) for n in x2]

#"O(n)"
x3 = range(1,10)
y3 = [n for n in x3]

#"O(n log n)"
x4 = range(1,10)
y4 = [(n * (math.log(n))) for n in x4]

#"O(n^2)"
x5 = range(1,8)
y5 = [n ** 2 for n in x5]

#"O(2^n)"
x6 = range(1,7)
y6 = [2 ** n for n in x6]

#"O(n!)"
x7 = range(1,4)
y7 = [math.factorial(n) for n in x7]

#"O(n^n)"
x8 = range(1,4)
y8 = [n ** n for n in x8]

plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.axhline(y=0, color="black")
plt.axvline(color="grey")

plt.plot(x1, y1)
plt.plot(x2, y2)
plt.plot(x3, y3)
plt.plot(x4, y4)
plt.plot(x5, y5)
plt.plot(x6, y6)
plt.plot(x7, y7)
plt.plot(x8, y8)

plt.legend(["$O(1)$", "$O(log (n))$", "$O(n)$", "$O(n log n)$", "$O(n^2)$", "$O(2^n)$", "$O(n!)$", "$O(n^n)$"])
```
