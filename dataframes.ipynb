# Copy the code from the lesson to create a dataframe full of student grades.

```python
import pandas as pd
import numpy as np

np.random.seed(123)

students = ['Sally', 'Jane', 'Suzie', 'Billy', 'Ada', 'John', 'Thomas',
            'Marie', 'Albert', 'Richard', 'Isaac', 'Alan']

# randomly generate scores for each student for each subject
# note that all the values need to have the same length here
math_grades = np.random.randint(low=60, high=100, size=len(students))
english_grades = np.random.randint(low=60, high=100, size=len(students))
reading_grades = np.random.randint(low=60, high=100, size=len(students))

df = pd.DataFrame({'name': students,
                   'math': math_grades,
                   'english': english_grades,
                   'reading': reading_grades})

type(df)
```

### Create a column named passing_english that indicates whether each student has a passing grade in reading.

```python
#bynitself, this creates a column of bool values for passing english
english_grades = df.english >= 70

#we create a new column with these values
df['passing_english'] = english_grades
df
```

### Sort the english grades by the passing_english column. How are duplicates handled?

```python
#we sort by passing_english and ascending false to see the passing grades first
#duplicates are ordered randomly?
df.sort_values(by='passing_english', ascending = False)
```

### Sort the english grades first by passing_english and then by student name. All the students that are failing english should be first, and within the students that are failing english they should be ordered alphabetically. The same should be true for the students passing english. (Hint: you can pass a list to the .sort_values method)

```python
#can sort by more than one column by using a list in by=[]
df = df.sort_values(by=['passing_english','name'])
df
```

### Sort the english grades first by passing_english, and then by the actual english grade, similar to how we did in the last step.

```python
#what comes first in the list is sorted first with prescedence
df = df.sort_values(by=['passing_english','english'])
df
```

### Calculate each students overall grade and add it as a column on the dataframe. The overall grade is the average of the math, english, and reading grades.

```python
total_grade = (df.math + df.english + df.reading) / 3
df['overall_grade'] = total_grade.astype(int)
df
```

# Load the mpg dataset. Read the documentation for the dataset and use it for the following questions:

### How many rows and columns are there?

### What are the data types of each column?

### Summarize the dataframe with .info and .describe

```python
print(mpg.describe())
print(mpg.info())
```

### Rename the cty column to city.

```python
#we rename with the function and the old name first, new name second
mpg = mpg.rename(columns={'cty': 'city'})
mpg
```

### Rename the hwy column to highway.

```python
#we rename with the old name first, and the new name second
mpg = mpg.rename(columns={'hwy': 'highway'})
mpg
```

### Do any cars have better city mileage than highway mileage?

```python
#no cars have better city than highway mpg
print(mpg[mpg.city >= mpg.highway])

#comparing the opposite shows the code does work
mpg[mpg.city < mpg.highway]
```

### Create a column named mileage_difference this column should contain the difference between highway and city mileage for each car.

```python
#we create a new column with the highway column minus the city column
mpg['mileage_difference'] = mpg.highway - mpg.city
mpg
```

### Which car (or cars) has the highest mileage difference?

```python
#count holds the amount of duplicate maximum values for mileage difference
count = mpg.mileage_difference.nlargest(1, keep='all').count()

#we sort by mileage difference ascending to get the highest values first, use head of hte count
mpg.sort_values(by='mileage_difference', ascending = False).head(count)
```

```python
#another way is to combine both conditionals into a subset DF
mpg[mpg.mileage_difference == mpg.mileage_difference.max()]
```

### Which compact class car has the lowest highway mileage? The best?

```python
#creates a dataframe of compact cars
compacts = mpg[mpg['class'] == "compact"]
#finds minimum highway mileage, and returns only the cars with the same value
compacts[compacts['highway'] ==  compacts['highway'].min()]
```

```python
#creates a dataframe of compact cars
compacts = mpg[mpg['class'] == "compact"]
#finds maximum highway mileage, and returns only the cars with the same value
compacts[compacts['highway'] ==  compacts['highway'].max()]
```

### Create a column named average_mileage that is the mean of the city and highway mileage.

```python
#calculates the average mileage by adding city and highway and dividing by two
average_mileage = (mpg.city + mpg.highway)/2
#create a new column with the variable above
mpg['average_mileage'] = average_mileage
```

### Which dodge car has the best average mileage? The worst?

```python
#finds how many cars are tied with the highest average mileage
count = mpg[mpg.manufacturer == 'dodge'].average_mileage.nlargest(1, keep='all').count()
#orders by best average mileage and shows only the top from the count above
mpg[mpg.manufacturer == 'dodge'].sort_values(by='average_mileage', ascending = False).head(count)
```

```python
#finds how many cars are tied with the lowest average mileage
count = mpg[mpg.manufacturer == 'dodge'].average_mileage.nsmallest(1, keep='all').count()
#orders by lowest average mileage and shows only the top from the count above
mpg[mpg.manufacturer == 'dodge'].sort_values(by='average_mileage').head(count)
```

# Load the Mammals dataset. Read the documentation for it, and use the data to answer these questions:

```python
#load the data and store in variable
mammals = data('Mammals')
#take a peak at the first five rows to check we loaded the data
mammals.head(5)
```

### How many rows and columns are there?

```python
#shape returns the number of rows and columns in a list
print('There are',mammals.shape, 'respective rows and columns in mammals.')
```

### What are the data types?

```python
#dtypes returns the data type for each column
mammals.dtypes
```

### Summarize the dataframe with .info and .describe

```python
print(mammals.info())
print(mammals.describe())
```

### What is the the weight of the fastest animal?

```python
#fastest orders the mammals by lowest spead, and holds only the last item - the fastest
fastest = mammals.sort_values(by='speed').tail(1)
#selects only the weight column of the fastest
fastest['weight']
```

### What is the overall percentage of specials?

```python
#value_counts counts each special true and false value
#normalize = true converts these counts into a ratio decimal
#multiplying by 100 gives us the percentage
mammals['specials'].value_counts(normalize=True) * 100
```

### How many animals are hoppers that are above the median speed? What percentage is this?

```python
#finds the median speed of all mammals in the DF
median_speed = mammals['speed'].median()
#creates a DF of only hoppers in the mammals DF
hoppers = mammals[mammals['hoppers'] == True]
#creates a subset DF of all hoppers with a speed greater than the median
hoppers = hoppers[hoppers['speed'] > median_speed]
hoppers
```
