# Use the iris database to answer the following quesitons:

```python
#loading the iris dataset into s variable
iris = sns.load_dataset('iris')

#downloading as csv
iris.to_csv('iris.csv')
```

```python
#displaying the first five rows to get a feel for the data :D
iris.head()
```

### 1. What does the distribution of petal lengths look like?

```python
#plotting the species as the x axis and petal length as the y
#setting the hue as the species alowws us to see trends in the specific species
sns.relplot(x='species', y='petal_length', data=iris, hue='species')

#adding title to plot
plt.title('Species vs. Petal Length')
```

```python
#the plot can also be represented with x and y switched
#this demonstrates without the hue as well
sns.relplot(x='petal_length', y='species', data=iris)

#adding title to plot
plt.title('Petal Length vs. Species')
```

### 2. Is there a correlation between petal length and petal width?

```python
#lmplot plots the data along with a line that best fits it
#we can see there is a correlation between length and width of petals
sns.lmplot(x='petal_length',y='petal_width', data=iris)
plt.title('Petal Length vs. Width')
```

### 3. Would it be reasonable to predict species based on sepal width and sepal length?

```python
#plotting relationship between species and sepal length
sns.scatterplot(x="sepal_length", y="species", data=iris, hue = 'species')
plt.suptitle("Sepal Length vs. Species", y=1, fontsize = 16)
```

```python
sns.scatterplot(x="sepal_width", y="species", data=iris, hue = 'species')
plt.suptitle("Sepal Width vs. Species", y=1, fontsize = 16)
```

```python
#lmplot plots data and regression model fits
#regression lines determine the strength and character of the relationship between x and y

#looking at the data split up by species
sns.lmplot(x="sepal_length", y="sepal_width", data=iris, col="species", hue='species', palette = 'deep')
plt.suptitle("Species Sepal Length vs. Width", y=1.1, fontsize = 16)

#looking at the data swith axis switched and regression line added
sns.lmplot(x="sepal_width", y="sepal_length", col="species", hue='species',data=iris, palette ="pastel")
plt.suptitle("Species Sepal Width vs. Length", y=1.1, fontsize = 16)
```

```python
#lmplot plots data and regression model fits
#using hue for species, we can see the correlation between the species and sepal size
#however, the distinction is not as clear between the orange and green data points
sns.lmplot(x='sepal_length',y='sepal_width', data=iris, hue='species', palette = 'deep')
#adds title
plt.title('Sepal Length vs. Width')

#same plot but with variables on different axis
sns.lmplot(x='sepal_width',y='sepal_length', data=iris, hue='species', palette = 'pastel')
#adds title
plt.title('Sepal Width vs. Length')
```

You could use sepal length and width to predict if the flower was Setosa or not, but there is not enough distinction between Versicolor and Virginica.

### 4. Which features would be best used to predict species?

```python
#pairplot plots pairwise relationships in a dataset
#using this, we can see the most linear data is with petal length vs petal width
sns.pairplot(iris, hue='species')
```

# 3. Load the swiss dataset and read it's documentation. Create visualizations to answer the following questions:

"Switzerland, in 1888, was entering a period known as the demographic transition; i.e., its fertility was beginning to fall from the high level typical of underdeveloped countries."
https://stat.ethz.ch/R-manual/R-patched/library/datasets/html/swiss.html

```python
#loads the dataset into a variables
swiss = data('swiss')

#takes a peek at the data
swiss.head()

data('swiss', show_doc = True)
```

### Create an attribute named is_catholic that holds a boolean value of whether or not the province is Catholic. (Choose a cutoff point for what constitutes catholic)

```python
#creates a new boolean column based on the amount of Catholic
#if at least a passing grade, is_catholic = True
swiss['is_catholic'] = swiss.Catholic >= 70
```

```python
#checking that new column is in the data
swiss.head()
```

### Does whether or not a province is Catholic influence fertility?

```python
#creating a function to determine if fertility rate high or low
def check_fertility(x):
    if x < 70:
        return 'low'
    else:
        return 'high'

#apllying the function to a new column in the DF
swiss['is_fertile'] = swiss.Fertility.apply(check_fertility)

#checking the column is updated and correct
swiss.head()
```

```python
#ploting with lmplot so see if a linear correlation is present
sns.lmplot(x = 'Fertility', y = 'Catholic', hue = 'is_fertile', data=swiss)

#there is no coreelation between being catholic and infant mortality
```

```python
#ploting with lmplot so see if a linear correlation is present
sns.lmplot(x = 'Catholic', y = 'Fertility', hue = 'is_fertile', data=swiss)

#there is no coreelation between being catholic and infant mortality
```

### What measure correlates most strongly with fertility?

```python
#visualizing with pairplot of all numeric variables
sns.pairplot(swiss, vars = ['Agriculture','Examination','Catholic','Infant.Mortality', 'Fertility'],hue = 'is_fertile',kind = 'reg', corner = True)
```

```python
#plots a pairplot of the 4 most prominent correlation of variables with fertility
top_4 = sns.pairplot(swiss, vars = ['Education','Examination', 'Agriculture', 'Fertility'], hue = 'is_fertile', kind = 'reg', corner = True)

#adds the title
top_4.fig.suptitle("Corellation of Fertility Between Examination and Agriculture", y=1.05)

#Examination shows the highest corelation
```

# 4. Using the chipotle dataset from the previous exercise, create a bar chart that shows the 4 most popular items and the revenue produced by each.

```python
#defines function to create a sql url using personal credentials
def get_db_url(db_name):
    
    #importing variables used to connect to sql database
    from env import host, username, password
    
    #creates the url and the function returns it
    url = f'mysql+pymysql://{username}:{password}@{host}/{db_name}'
    return (url)

orders = pd.read_sql('''SELECT * FROM orders''',get_db_url('chipotle'))
```

```python
#taking a peek at the data
orders.head()
```

```python
#converting the price column to a float to be able to use aggregate functions of
orders['item_price'] = orders.item_price.str.replace('$','').astype(float)
```

```python
#checking the column was updated
orders.head()
```

```python
#creates a new dataframe based on the item name and the sum of prices for each item
item_revenue = pd.DataFrame(orders.groupby('item_name')['item_price'].sum())
```

```python
#updating the DF to sort by highest revenue first
item_revenue = item_revenue.sort_values(by='item_price', ascending=False)
```

```python
#creates a bar plot of the first four rows in the new DF created above
item_revenue.head(4).plot.bar()

#adds a plot title
plt.title('Four Highest Items from Chipotle by Revenue', fontsize=16)

#adds the x axis with formating on the font size and orientation for clarity
plt.xlabel('Item Name', fontsize=16)
plt.xticks(rotation='horizontal', fontsize=14)

#adds the y axis with formating on the font size and orientation for clarity
plt.ylabel('Revenue in $', fontsize=16)
plt.yticks(fontsize=14)
```

# 5. Load the sleepstudy data and read it's documentation. Use seaborn to create a line chart of all the individual subject's reaction times and a more prominant line showing the average change in reaction time.

```python
#loads the dataset from py into a variable
sleepstudy = data('sleepstudy')
```

```python
#checks the documentation
data('sleepstudy', show_doc=True)
```

```python
#taking a peek at the data
sleepstudy.head()
```

```python
#creates a variable to hold the averages of the data frame by Days
avg_reaction = sleepstudy.groupby('Days').mean()

#we drop the average subject because it is not needed to know the average reaction times
avg_reaction.drop(columns=['Subject'])

#convert reaction as an int to take away long number decimals
avg_reaction['Reaction'] = avg_reaction.Reaction.astype(int)
```

```python
#checks the new DF is correct
avg_reaction.head()
```

```python
#plt.scatter(x = avg_reaction.index, y = avg_reaction.Reaction )

#creating x and y variables based on the average reaction time dataframe
x = avg_reaction.index
y = avg_reaction.Reaction

#plots the average line
plt.plot(x,y, linewidth=6, linestyle = '--')

#plots all sleepstudy subjects by reaction over time(days)
sns.lineplot('Days', 'Reaction', hue='Subject', data=sleepstudy, palette="Paired")

#formats the plot legend and figure size
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0)
plt.figure(figsize=(12, 10))
plt.show()
```
