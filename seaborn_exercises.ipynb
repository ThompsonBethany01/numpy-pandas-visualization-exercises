# Use the iris database to answer the following quesitons:

```python
#loading the iris dataset into s variable
iris = sns.load_dataset('iris')

#downloading as csv
iris.to_csv('iris.csv')
```

```python
#displaying the first five rows to get a feel for the data :D
iris.head()
```

### 1. What does the distribution of petal lengths look like?

```python
#plotting the species as the x axis and petal length as the y
#setting the hue as the species alowws us to see trends in the specific species
sns.relplot(x='species', y='petal_length', data=iris, hue='species')

#adding title to plot
plt.title('Species vs. Petal Length')
```

```python
#the plot can also be represented with x and y switched
#this demonstrates without the hue as well
sns.relplot(x='petal_length', y='species', data=iris)

#adding title to plot
plt.title('Petal Length vs. Species')
```

### 2. Is there a correlation between petal length and petal width?

```python
#lmplot plots the data along with a line that best fits it
#we can see there is a correlation between length and width of petals
sns.lmplot(x='petal_length',y='petal_width', data=iris)
plt.title('Petal Length vs. Width')
```

### 3. Would it be reasonable to predict species based on sepal width and sepal length?

```python
#lmplot plots the data along with a line that best fits it
#using hue for species, we can see hte correlation between the species and petal size
#however, the distinction is not as clear between the orange and green data points
sns.lmplot(x='petal_length',y='petal_width', data=iris, hue='species')
plt.title('Petal Length vs. Width')
```

```python
#looking at the data split up by species
sns.relplot(x="petal_length", y="petal_width", data=iris, col="species", hue='species')
plt.suptitle("Species Petal Length vs. Width", y=1.1, fontsize = 16)
```

```python
#looking at the data split up and ading a line for reference
sns.lmplot(x="petal_length", y="petal_width", col="species", hue='species',data=iris)
plt.suptitle("Species Petal Length vs. Width", y=1.1, fontsize = 16)
```

### 4. Which features would be best used to predict species?

```python
#pairplot plots pairwise relationships in a dataset
#using this, we can see the most linear data is with petal length vs petal width
sns.pairplot(iris, hue='species')
```

### Create an attribute named is_catholic that holds a boolean value of whether or not the province is Catholic. (Choose a cutoff point for what constitutes catholic)

```python
#creates a new boolean column based on the amount of Catholic
#if at least a passing grade, is_catholic = True
swiss['is_catholic'] = swiss.Catholic >= 70
```

```python
#checking that new column is in the data
swiss.head()
```

### Does whether or not a province is Catholic influence fertility?

```python
#ploting with lmplot so see if a linear correlation is present
sns.lmplot(x = 'Catholic', y = 'Infant.Mortality', hue = 'is_catholic', data=swiss)

#there is no coreelation between being catholic and infant mortality
```

### What measure correlates most strongly with fertility?

```python
#plots a pairplot of the most prominent correlation of variables with fertility
plot = sns.pairplot(swiss, vars = ['Education','Examination', 'Agriculture', 'Fertility'], kind = 'reg', corner = True)

#adds the title
plot.fig.suptitle("Corellation of Fertility Between Examination and Agriculture", y=1.05)

#Examination shows the highest corelation
```

# 4. Using the chipotle dataset from the previous exercise, create a bar chart that shows the 4 most popular items and the revenue produced by each.

```python
#defines function to create a sql url using personal credentials
def get_db_url(db_name):
    
    #importing variables used to connect to sql database
    from env import host, username, password
    
    #creates the url and the function returns it
    url = f'mysql+pymysql://{username}:{password}@{host}/{db_name}'
    return (url)

orders = pd.read_sql('''SELECT * FROM orders''',get_db_url('chipotle'))
```

```python
orders.head()
```

```python
orders['item_price'] = orders.item_price.str.replace('$','').astype(float)
```

```python
orders.head()
```

```python
item_revenue = pd.DataFrame(orders.groupby('item_name')['item_price'].sum())
```

```python
item_revenue = item_revenue.sort_values(by='item_price', ascending=False)
```

```python
item_revenue.head(4).plot.bar()
plt.title('Four Highest Items from Chipotle by Revenue', fontsize=16)
plt.xlabel('Item Name', fontsize=16)
plt.xticks(rotation='horizontal', fontsize=14)
plt.ylabel('Revenue in $', fontsize=16)
plt.yticks(fontsize=14)
plt.margins(0.2)
```
