# 1. Use matplotlib to plot the following equation:

y = x^2 − x + 2

You'll need to write the code that generates the x and y points.
Add an anotation for the point 0, 0, the origin.

```python
import matplotlib.pyplot as plt
```

```python
x = range(-100, 101, 1)
y = [((n ** 2) - n + 2) for n in x]

plt.plot(x,y)
plt.title('The Quadratic Formula')
plt.xlabel('x')
plt.ylabel('$n ^ 2 - n + 2$')
plt.axhline(y=0, color="black")
plt.axvline(color="black")
plt.annotate('Origin (0,0)', xy=(0, 0), xytext=(-20,1500), fontsize=12, color='blue', arrowprops={'facecolor': 'blue'})
plt.show()
```

function for semicircle found here

```python
import matplotlib.pyplot as plt
from matplotlib import pyplot, transforms
import numpy as np

def semicircle(center_x, center_y, radius, stepsize=0.1):
    """
    generates coordinates for a semicircle, centered at center_x, center_y
    """        

    x = np.arange(center_x, center_x+radius+stepsize, stepsize)
    y = np.sqrt(radius**2 - x**2)

    # since each x value has two corresponding y-values, duplicate x-axis.
    # [::-1] is required to have the correct order of elements for plt.plot. 
    x = np.concatenate([x,x[::-1]])

    # concatenate y and flipped y. 
    y = np.concatenate([y,-y[::-1]])

    return x, y + center_y

def top_semicircle_for_e(center_x, center_y, radius, stepsize=0.1):
    """
    generates coordinates for a semicircle, centered at center_x, center_y
    """        

    x = np.arange(center_x, center_x+radius+stepsize, stepsize)
    y = np.sqrt(radius**2 - x**2)

    # since each x value has two corresponding y-values, duplicate x-axis.
    # [::-1] is required to have the correct order of elements for plt.plot. 
    x = np.concatenate([x,-x[::-1]])

    # concatenate y and flipped y. 
    y = np.concatenate([y,y[::-1]])

    return x + 40, y + center_y - 5

def bottom_semicircle_for_e(center_x, center_y, radius, stepsize=0.1):
    """
    generates coordinates for a semicircle, centered at center_x, center_y
    """        

    x = np.arange(center_x, center_x+radius+stepsize, stepsize)
    y = np.sqrt(radius**2 - x**2)

    # since each x value has two corresponding y-values, duplicate x-axis.
    # [::-1] is required to have the correct order of elements for plt.plot. 
    x = np.concatenate([x,-x[::-1]])

    # concatenate y and flipped y. 
    y = np.concatenate([y,-y[::-1]])

    return x + 40, y + center_y - 5

def curve_for_h(center_x, center_y, radius, stepsize=0.1):
    """
    generates coordinates for a semicircle, centered at center_x, center_y
    """        

    x = np.arange(center_x, center_x+radius+stepsize, stepsize)
    y = np.sqrt(radius**2 - x**2)

    # since each x value has two corresponding y-values, duplicate x-axis.
    # [::-1] is required to have the correct order of elements for plt.plot. 
    x = np.concatenate([x,x[::-1]])

    # concatenate y and flipped y. 
    y = np.concatenate([y,y[::-1]])

    return x + 96, y + center_y - 15

def semicircle_for_a(center_x, center_y, radius, stepsize=0.1):
    """
    generates coordinates for a semicircle, centered at center_x, center_y
    """        

    x = np.arange(center_x, center_x+radius+stepsize, stepsize)
    y = np.sqrt(radius**2 - x**2)

    # since each x value has two corresponding y-values, duplicate x-axis.
    # [::-1] is required to have the correct order of elements for plt.plot. 
    x = np.concatenate([x,-x[::-1]])

    # concatenate y and flipped y. 
    y = np.concatenate([y,y[::-1]])

    return x + 14, y + center_y + 20

def curve_for_n(center_x, center_y, radius, stepsize=0.1):
    """
    generates coordinates for a semicircle, centered at center_x, center_y
    """        

    x = np.arange(center_x, center_x+radius+stepsize, stepsize)
    y = np.sqrt(radius**2 - x**2)

    # since each x value has two corresponding y-values, duplicate x-axis.
    # [::-1] is required to have the correct order of elements for plt.plot. 
    x = np.concatenate([x,x[::-1]])

    # concatenate y and flipped y. 
    y = np.concatenate([y,y[::-1]])

    return x + 150, y + center_y - 18

def curve_for_y(center_x, center_y, radius, stepsize=0.1):
    """
    generates coordinates for a semicircle, centered at center_x, center_y
    """        

    x = np.arange(center_x, center_x+radius+stepsize, stepsize)
    y = np.sqrt(radius**2 - x**2)

    # since each x value has two corresponding y-values, duplicate x-axis.
    # [::-1] is required to have the correct order of elements for plt.plot. 
    x = np.concatenate([x,x[::-1]])

    # concatenate y and flipped y. 
    y = np.concatenate([y,y[::-1]])

    return x - 205, y + center_y - 48
```

```python
#line for spacing reference
#x1 = 100, 100, 100
#y1 = 0, 30, 60

#first vertical line in B
x2 = 0, 0, 0
y2 = 0, 30, 60

#first semicircle in B
x3,y3 = semicircle(0,50,10, 0.1)

#second semicircle in B
x4,y4 = semicircle(0,20,20, 0.1)

#top semicircle in e
x5,y5 = top_semicircle_for_e(0,20,15, 0.1)

#top semicircle in e
x6,y6 = bottom_semicircle_for_e(0,20,15, 0.1)

#vertical line in t
x7 = 75, 75, 75
y7 = 0, 2, 50

#horizontal line of t
x8 = 60, 70, 90
y8 = 30, 30, 30

#vertical line in h
x9 = 95, 95, 95
y9 = 1, 2, 50

#curve in h
x10,y10 = curve_for_h(0,20,25, 0.1)

#bottom horizontal line of h
x11 = 121.25, 121.25, 121.25
y11 = 1, 3, 5

#semicircle for a
x12, y12 = semicircle_for_a(0, -160, 13, .1)
#sets transformation to rotate semicircle
base = pyplot.gca().transData
rot = transforms.Affine2D().rotate_deg(90)

#little vertical line for a
x13 = 140, 140, 140
y13 = 0, 2, 29

#vertical line in n
x14 = 150, 150, 150
y14 = 0, 2, 29

#same curve of h used for n, function repeated to adjust axis
x15,y15 = curve_for_n(0,20,25, 0.1)

#bottom horizontal line of n
x16 = 175, 175, 175
y16 = 0, 3, 5

#vertical line in y
x17 = 206, 206, 206
y17 = -20, 2, 28.25

#same curve of h used for y, function repeated to adjust axis
x18, y18 = curve_for_y(0,20,28, 0.1)

#rotation for y
ybase = pyplot.gca().transData
yrot = transforms.Affine2D().rotate_deg(180)

#plt.plot(x1, y1)
plt.plot(x2, y2, c='blueviolet', linewidth=4)
plt.plot(x3, y3, c='blueviolet', linewidth=4)
plt.plot(x4, y4, c='blueviolet', linewidth=4)
plt.plot(x5, y5, c='mediumpurple', linewidth=4)
plt.plot(x6, y6, c='mediumpurple', linewidth=4)
plt.plot(x7, y7, c='thistle', linewidth=4)
plt.plot(x8, y8, c='thistle', linewidth=4)
plt.plot(x9, y9, c='lightpink', linewidth=4)
plt.plot(x10, y10, c='lightpink', linewidth=4)
plt.plot(x11, y11, c='lightpink', linewidth=4)
plt.plot(x12, y12, linewidth=4, c='palevioletred', transform= rot + base)
plt.plot(x13, y13, linewidth=4, c='palevioletred')
plt.plot(x14, y14, linewidth=4, c='hotpink')
plt.plot(x15, y15, linewidth=4, c='hotpink')
plt.plot(x16,y16, linewidth=4, c='hotpink')
plt.plot(x17, y17, linewidth=4, c='violet')
plt.plot(x18, y18, linewidth=4, c='violet', transform= yrot + ybase)

plt.show()
```
